**Define y(dependent or lable or target variable) and x (independent or features or attribute Variable)**

```python
y = df['diabetes']
```

```python
y.shape
```

```python
y
```

```python
x = df[['pregnancies','glucose','diastolic','triceps','insulin','bmi','dpf','age']]
```

**or use .drop function to define x**

```python
x = df.drop('diabetes', axis=1)
```

```python
x.shape
```

```python
x
```

**Get x Variable Standerised**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x = mn.fit_transform(x)
```

```python
x
```

**Get Train_Test_Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, stratify= y, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Get Model Train**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

**Get Model Prediction**

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get Probability of Each Predicted Class**

```python
lr.predict_proba(x_test)
```

**Get Model Evalution**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test,y_pred))
```
